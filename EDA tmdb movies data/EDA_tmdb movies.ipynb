## 2) What is the average number of words in movie titles between the years 2000-2005?


```python
# Subset release year between 2000 and 2015
df_subset = df[(df["release_year"] > 2000) & (df["release_year"] < 2005)]
```

```python
df_subset['release_year']
```

```python
# Number of words in Movie titles between year 2000-2005
words = df_subset['original_title'].str.split().apply(len)
words
```

```python
# Dataframe for better understanding
df_final = pd.DataFrame({
 'original_title' : df_subset['original_title'] ,
 'No. of words' : words})
df_final
```

```python
df_final.shape
```

```python
# Average number of words in movie titles between 2000-2005:
avg = df_final['No. of words'].mean()
avg
```

### Average number of words in movie titles between the years 2000-2005 is 2.87
